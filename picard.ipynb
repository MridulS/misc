```python
from sympy import exp, Function
from sympy import symbols, integrate, Subs
import matplotlib
from sympy.plotting import plot
%matplotlib inline
```

```python
def picard(f, a, c, iterations):
    if iterations == 0:
        return c
    if iterations == 1:
        x0 = lambda t: c + integrate(Subs(f, (t, x), (s, c)), (s, a, t)).doit()
        return x0(t)
    for i in range(iterations):
        x_old = lambda s: picard(f, a, c, iterations-1).subs(t, s)
        x0 = lambda t: c + integrate(Subs(f, (t, x), (s, x_old(s))), (s, a, t)).doit()
    return x0(t)
```

```python
a = 0; c = 2
```

```python
(s, t, x) = symbols('s t x')
f = Function('f')('x, t')
f = x
```

```python
picard(f, a, c, 3)
```

```python
plot(picard(f, a, c, 8))
```

```python
plot(exp(t))  # This is supposed to be the answer
```

```python
plot(picard(f, a, c, 20))
```

We can see that as we increase the number of iterations we get closer to the actual answer which is e^t in this case for the IVP x' = f(x, t) where f(x, t) = x and x(0) = 2.
